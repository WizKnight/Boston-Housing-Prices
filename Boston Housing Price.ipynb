# Why do you standardise dataset in linear regression?
**Ans.** Because internally we use gradient decent, our main aim is to come close to global minima and to come close to global minima we need to make sure that all our independent feature unit should be in the same scale, and because of that conversions will happen very quickly.

```python
## Standardising the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
X_test
```

# Model Trining

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
```

```python
lreg.fit(X_train, y_train)
```

```python
## print the coefficients and the intercepts
print(lreg.coef_)
```

```python
print(lreg.intercept_)
```

```python
## On which parameter the model has been trained
lreg.get_params()
```

```python
### Prediction with test data
reg_pred = lreg.predict(X_test)
```

```python
reg_pred
```

```python
## For X_test, y_test is the true value
```

# Assumptions

```python
## Plot a scatter plot for the prediction

plt.scatter(y_test, reg_pred) #y_test is truth value and reg_pred are predicted values
```

```python
## Errors w.r.t y_test and reg_pred
residuals = y_test - reg_pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

# R square and adjusted R square

Formula:

**R^2 = 1 - SSR/SST**

R^2 = coefficient of determination, SSR = sum of squares of residuals, SST = total sum of squares

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

**Adjusted R2 = 1  [(1-R2)*(n-1)/(n-k-1)]**

Where:
* R2: The R2 of the model
* n: The number of observations
* k: The number of predictor variables

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
